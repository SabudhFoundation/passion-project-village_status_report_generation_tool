Analyze missing values and their percentages for each column.

### 4.1 Drop Columns with High Missing Percentage
Remove columns with more than 60% missing values.


```python
threshold = 60
cols_to_drop = missing_report[missing_report["Missing_Percent"] > threshold].index.tolist()
school_data_cleaned = school_data.drop(columns=cols_to_drop)
```

## 5.1 Correct Negative Values
Replace negative counts with absolute values (typos).


```python
school_data_cleaned["Children_Village_Age0_3"] = school_data_cleaned["Children_Village_Age0_3"].abs()
school_data_cleaned["Students_Grade3_Boys"] = school_data_cleaned["Students_Grade3_Boys"].abs()
school_data_cleaned["Students_Grade4_Boys"] = school_data_cleaned["Students_Grade4_Boys"].abs()
```

## 6. Handling UDISE Codes
Check for duplicates and conflicts in UDISE codes.

```python
# Count duplicates
frequency = school_data_cleaned['UDISE_Code'].value_counts()

# Rows with duplicated UDISE codes
duplicates = school_data_cleaned[school_data_cleaned['UDISE_Code'].duplicated()]
```

```python
udise_multi_names = school_data_cleaned.groupby('UDISE_Code')['School_Name'].nunique()
udise_multi_names = udise_multi_names[udise_multi_names > 1].index
df_conflict = school_data_cleaned[school_data_cleaned['UDISE_Code'].isin(udise_multi_names)]
df_conflict = df_conflict.sort_values(by='UDISE_Code')
df_conflict
```

```python
# Correct known conflicts
school_data_cleaned.loc[239, "School_Name"] = "GPS PIPLI MEHRAJ"
school_data_cleaned.loc[467, "UDISE_Code"] = "03070607301"
```

## 7. Convert Assessment Dates
Convert `Assessment_Date` to datetime for proper calculations.


```python
school_data_cleaned['Assessment_Date'] = pd.to_datetime(school_data_cleaned['Assessment_Date'])
```

## 8. Identify Close Assessment Dates

*  For UDISE codes with multiple assessments, records with small date differences i.e., <=90 days were identified.
* For these records, the absolute difference in total student count across grades was calculated to assess potential inconsistencies.


```python
data_frequency = school_data_cleaned.sort_values(by=['UDISE_Code', 'Assessment_Date'])
data_frequency['DateDiff'] = data_frequency.groupby('UDISE_Code')['Assessment_Date'].diff().dt.days

same_or_close = data_frequency[(data_frequency['DateDiff'] == 0) | (data_frequency['DateDiff'] <= 90)]
same_or_close 
```

## 8.1 Compute Absolute Student Differences
Filter rows based on absolute difference in total student counts.

```python
udise_list = same_or_close['UDISE_Code'].unique()

filtered_data = data_frequency[data_frequency['UDISE_Code'].isin(udise_list)]
filtered_data['Abs_student_diff'] = abs(data_frequency['Students_Total_GPS'] - (data_frequency['Students_PrePrimary_Girls']+ data_frequency['Students_PrePrimary_Boys']+ data_frequency['Students_Grade1_Girls'] + data_frequency['Students_Grade1_Boys'] + data_frequency['Students_Grade2_Girls'] + data_frequency['Students_Grade2_Boys'] + data_frequency['Students_Grade3_Girls'] + data_frequency['Students_Grade3_Boys'] + data_frequency['Students_Grade4_Girls'] + data_frequency['Students_Grade4_Boys'] + data_frequency['Students_Grade5_Girls'] + data_frequency['Students_Grade5_Boys']))
filtered_data = filtered_data.sort_values(by = ['UDISE_Code','Assessment_Date'])
filtered_data['Assessment_Date'] = pd.to_datetime(filtered_data['Assessment_Date'])
filtered_data = filtered_data[filtered_data['Assessment_Date'].dt.year >= 2025]
filtered_data[['UDISE_Code','School_Name','Abs_student_diff','Assessment_Date','DateDiff']]
```

```python
# Drop rows with maximum absolute student difference per UDISE code
indices_to_drop = filtered_data.groupby('UDISE_Code')['Abs_student_diff'].idxmax()
school_data_cleaned.drop(indices_to_drop, inplace=True)
```

## 9. Save Cleaned Dataset
Export the final cleaned dataset to Excel.


```python
school_data_cleaned.to_excel("school_data.xlsx", index=False)
```
